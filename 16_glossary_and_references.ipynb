## Glossary

- **Pretext task** — a fake task whose target is taken from the unlabeled data itself
  (mask-and-predict, reconstruct, predict latent); the basis of self-supervision.
- **Rank `r`** — the inner dimension of a LoRA update `ΔW = BA`; controls capacity and
  trainable-parameter count (linear in `r`).
- **`alpha` / scaling** — LoRA's update is scaled by `alpha / r` (or `alpha / √r` in rsLoRA).
- **NF4** — 4-bit NormalFloat, the quantization datatype QLoRA stores the frozen base in.
- **Double quantization** — quantizing the quantization constants too (QLoRA), saving more memory.
- **Paged optimizer** — spills optimizer state to CPU on memory spikes instead of OOM-ing.
- **Catastrophic forgetting** — loss of general ability when a model is specialized too narrowly.
- **Completion-only / response-only loss** — masking prompt tokens (label `-100`) so loss is
  computed only on the answer; what makes instruction tuning "done right".
- **Virtual tokens** — trainable soft-prompt embeddings prepended to the input (prompt tuning).
- **Reference model** — a frozen copy used by preference methods to anchor how far the policy
  may drift; `beta` controls that strength. DPO/ORPO can be reference-free.
- **Reward model (RM)** — a model that scores a response; trained by `RewardTrainer`, consumed
  by PPO / best-of-N.
- **On-policy** — training on the model's *own* current samples (GRPO, on-policy GKD), vs.
  offline/fixed data.
